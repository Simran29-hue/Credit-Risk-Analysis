```python
# check how many nulls are still exist in each column

missing = df.isnull() .sum()

print ('Missing values in each column')
print (missing)
```

```python
# Count how many customers defaulted vs did not defaulted
# loan_status = 1, Defaulted, 0 = not defaulted

plt.figure (figsize = (8,5))
sns.countplot (x ='loan_status', data=df, palette= ['#2ecc71', '#e74c3c'])
plt.title ('Default vs Non-default customers', fontsize= 16)
plt.xlabel ('loan_status = (0 = Non Defaulter, 1 = Defaulter)')
plt.ylabel ('Number of customers')
plt.xticks ([0,1], ['Non Defaulter', 'Defaulter'])
plt.show()
```

```python
# Check Default Rate Percentage

total= len(df)
defaulters = df ['loan_status'].sum()
default_rate = (defaulters / total) * 100

print(f"Total Loans: {total}")
print(f"Total Defaulters: {defaulters}")
print(f"Default Rate: {default_rate:.2f}%")
```
